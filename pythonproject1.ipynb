```python
import matplotlib.pyplot as plt
import yfinance as yf
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Closing Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()
# Extract GameStop stock data using yfinance
gme_data = yf.download("GME", start="2022-01-01", end="2023-01-01")
# Reset the index
gme_data_reset = gme_data.reset_index()
# Plot the GameStop stock graph
make_graph(gme_data_reset, title='GameStop Stock Price')
```

```python
import matplotlib.pyplot as plt
import yfinance as yf

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Closing Price', color='green')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Extract Tesla stock data using yfinance
tesla_data = yf.download("TSLA", start="2022-01-01", end="2023-01-01")

# Reset the index
tesla_data_reset = tesla_data.reset_index()

# Plot the Tesla stock graph
make_graph(tesla_data_reset, title='Tesla Stock Price')

```
